```python
#import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## search for target protein

```python
# target search for coronavirus
target = new_client.target
target_query = target.search('serotonin')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#choose bioactivity data for coronavirus 3C-like proteinase
selected_target = targets.target_chembl_id[4]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.to_csv('bioactivity_data.csv', index=False)
```

# Data pre-processing

```python
#check csv
check_df2_csv = pd.read_csv('/Users/verenakarina/Desktop/Project/Untitled Folder 1/bioactivity_data.csv')
check_df2_csv
```

```python
#handling missing data
df2 = df[df.standard_value.notna()]
pd.set_option("display.max_columns", None)
df2
```

```python
compound_classification = []
for i in df2.standard_value:
    if float(i) <= 1000:
        compound_classification.append("active")
    elif float(i) >= 10000:
        compound_classification.append("inactive")
    else:
        compound_classification.append("intermediate")
```

### combine necessary data

```python
select = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[select]
```

```python
 pd.concat([df3, pd.Series(compound_classification)], axis=1, names = ['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'compound_classification'])
```

```python
df3.to_csv('bioactiviey_preprocessed_data.csv', index=False)
```
