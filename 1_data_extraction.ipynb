```python
#import necessary libraries
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
```

## search for target protein

```python
# target search for coronavirus
target = new_client.target
target_query = target.search('serotonin')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#choose bioactivity data for serotonin receptor for homo sapiens
selected_target = targets.target_chembl_id[2]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
# pd.set_option("display.max_columns", None)
# pd.set_option("display.max_rows", None)
df.head(10)
```

```python
df.standard_type.unique()
```

```python
import os

path = 'csv/'
df.to_csv(os.path.join(path,'bioactivity_data.csv'),index=False)
```

# Data pre-processing

```python
file = pd.read_csv('csv/bioactivity_data.csv')
df2 = pd.DataFrame(file)
df2
```

```python
#take necessary data
attributes = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df2 = df[attributes]
df2
```

```python
df2.dtypes
```

```python
df2.shape
```

```python
#remove missing values
df2.dropna(inplace=True)
df2.reset_index(drop=True, inplace=True)
df2.shape
```

```python
compound_classification = []
for i in df2.standard_value:
    if float(i) <= 1000:
        compound_classification.append("active")
    elif float(i) >= 10000:
        compound_classification.append("inactive")
    else:
        compound_classification.append("intermediate")
compound_classification = pd.DataFrame(compound_classification, columns = ['compound_classification'])
compound_classification
```

### combine necessary data

```python
preprocessed_data =  pd.concat([df2, compound_classification], axis=1, names = ['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'compound_classification'])
preprocessed_data.shape
```

```python
preprocessed_data
```

```python
path = 'csv/'
preprocessed_data.to_csv(os.path.join(path,'bioactivity_preprocessed_data.csv'),index=False)
```
